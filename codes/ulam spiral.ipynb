Importing necessary libraries

```python
import math
import numpy as np
import matplotlib.pyplot as plt
import seaborn
```

```python
def spiral(n):
    """
    Generates coordinates in a spiral pattern for a given number n.

    Parameters:
    n (int): The number for which coordinates are to be generated.

    Returns:
    list: A list containing the x and y coordinates of the given number in the spiral pattern.
    """
    k = math.ceil((math.sqrt(n) - 1) / 2)
    t = (2 * k + 1) 
    m = math.pow(t, 2)
    t = t - 1
    if n >= m - t:
        return [k - (m - n), -k]
    
    else:
        m = m - t
    
    if n >= m - t:
       return  [-k, -k + (m - n)]
    else :
        m = m - t
    
    if n >= m - t:
        return  [-k + (m - n), k];
    
    else:
        return [k,k - (m - n - t)];
    
```

```python
spiral(4)
```

```python
semi_primes = open("semiprimes under 10^3.txt")
```

```python
numbers = []
```

```python
for i in semi_primes:
    numbers.append(int(i.strip(" ")))
length = len(numbers)
print(length)
```

```python
xs = []
ys = []
```

```python
count = 0
for i in range(length):
    count+=1
    coords = spiral(numbers[i])
    xs.append(int(coords[0]))
    ys.append(int(coords[1]))
print(count)
```

```python
semi_primes.close()
```

Visualization of numbers on a grid

```python
plt.figure(figsize=(20,20))
plt.tick_params(axis='x', labelsize=20)
plt.tick_params(axis='y', labelsize=20)
plt.scatter(xs,ys,s=10)
plt.grid(visible=True,which="both")
fig1 = plt.gcf()
plt.show()
fig1.savefig('semiprimes under 10^3.pdf', dpi=100)
fig1.savefig('semiprimes under 10^3.png', dpi=100)
fig1.savefig('semiprimes under 10^3.jpg', dpi=100)
```
